```python
import cv2
import os
from google.colab import drive

# Mount Google Drive (optional, for persistent storage)
drive.mount('/content/drive')

# Function to extract PNG frames from a video
def extract_frames(video_path, output_dir, frame_size=(256, 256), max_frames=None):
    """
    Extracts PNG frames from a video and saves them to output_dir.

    Args:
        video_path (str): Path to the input video file (e.g., .mp4).
        output_dir (str): Directory to save PNG frames.
        frame_size (tuple): Desired frame size (width, height), default 256x256 per DVC paper.
        max_frames (int): Optional limit on number of frames to extract.
    """
    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    # Open the video
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"Error: Could not open video {video_path}")
        return

    frame_count = 0
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret or (max_frames and frame_count >= max_frames):
            break

        # Resize frame to match DVC's training resolution (256x256)
        frame = cv2.resize(frame, frame_size, interpolation=cv2.INTER_AREA)

        # Save as PNG (DVC uses PNG inputs, e.g., im003.png)
        frame_path = os.path.join(output_dir, f"frame_{frame_count:03d}.png")
        cv2.imwrite(frame_path, frame)
        print(f"Saved {frame_path}")
        frame_count += 1

    cap.release()
    print(f"Extracted {frame_count} frames from {video_path}")

# # Example usage
video_path = '/content/drive/MyDrive/bechic2025_7414308146958634272.mp4'  # Replace with your video path
output_dir = '/content/frames'  # Temporary Colab directory
extract_frames(video_path, output_dir, frame_size=(256, 256))  # Extract 10 frames
```

```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
from imageio import imread
from
from DVC.TestDemo.VideoCodec.Decoder import decode_frame


def extract_frames(video_path, frames_dir):
    """Extracts frames from the video and saves them as images."""
    os.makedirs(frames_dir, exist_ok=True)
    cap = cv2.VideoCapture(video_path)
    frame_count = 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        frame_path = os.path.join(frames_dir, f'frame_{frame_count:04d}.png')
        cv2.imwrite(frame_path, frame)
        frame_count += 1

    cap.release()
    print(f"âœ… Extracted {frame_count} frames to {frames_dir}")
    return frame_count

def encode_frames(frames_dir, encoder_model, output_dir):
    """Encodes the extracted frames using the given encoder model."""
    os.makedirs(output_dir, exist_ok=True)
    frames = sorted(os.listdir(frames_dir))

    for i in range(1, len(frames)):
        input_frame = os.path.join(frames_dir, frames[i])
        refer_frame = os.path.join(frames_dir, frames[i - 1])
        encode_frame(encoder_model, input_frame, refer_frame, output_dir)

    print(f'âœ… Encoding complete. Compressed data stored in {output_dir}')

def decode_frames(frames_dir, decoder_model, encoded_dir, decoded_dir):
    """Decodes the compressed frames back into images."""
    os.makedirs(decoded_dir, exist_ok=True)
    frames = sorted(os.listdir(frames_dir))

    for i in range(1, len(frames)):
        refer_frame = os.path.join(frames_dir, frames[i - 1])
        decode_frame(decoder_model, refer_frame, encoded_dir, decoded_dir)

    print(f'âœ… Decoding complete. Decoded frames saved in {decoded_dir}')

def reconstruct_video(decoded_dir, output_video, fps=30):
    """Reconstructs the video from the decoded frames."""
    frames = sorted(os.listdir(decoded_dir))
    if not frames:
        print("âŒ No decoded frames found!")
        return

    sample_frame = cv2.imread(os.path.join(decoded_dir, frames[0]))
    height, width, _ = sample_frame.shape

    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_video, fourcc, fps, (width, height))

    for frame in frames:
        frame_path = os.path.join(decoded_dir, frame)
        img = cv2.imread(frame_path)
        out.write(img)

    out.release()
    print(f"âœ… Reconstructed video saved as {output_video}")

def main(video_path, encoder_model_path, decoder_model_path, output_dir):
    """Main function for video encoding and decoding."""
    frames_dir = os.path.join(output_dir, 'frames')
    encoded_dir = os.path.join(output_dir, 'encoded')
    decoded_dir = os.path.join(output_dir, 'decoded')
    reconstructed_video = os.path.join(output_dir, 'reconstructed_video.mp4')

    # Step 1: Extract frames from the input video
    extract_frames(video_path, frames_dir)

    # Step 2: Load your encoder and decoder models (example paths)
    encoder_model = encoder_model_path
    decoder_model = decoder_model_path

    # Step 3: Encode frames
    encode_frames(frames_dir, encoder_model, encoded_dir)

    # Step 4: Decode frames
    decode_frames(frames_dir, decoder_model, encoded_dir, decoded_dir)

    # Step 5: Reconstruct the video from decoded frames
    reconstruct_video(decoded_dir, reconstructed_video)


if __name__ == "__main__":
    # Example paths - replace with your own paths
    video_path = '/content/drive/MyDrive/bechic2025_7414308146958634272.mp4'
    encoder_model_path = '/content/DVC/TestDemo/VideoCodec/model/L256/frozen_model_E.pb'
    decoder_model_path = '/content/DVC/TestDemo/VideoCodec/model/L256/frozen_model_D.pb'
    output_dir = './output'

    main(video_path, encoder_model_path, decoder_model_path, output_dir)
```
